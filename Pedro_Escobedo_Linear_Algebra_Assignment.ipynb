## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
h = np.array([[12,11],[7,10]])
j = np.array([[0,1,2],[7,10,4],[3,2,0]])

det_h = np.linalg.det(h)
det_j = np.linalg.det(j)

print(det_h)
print(det_j)
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
inv_h = np.linalg.inv(h)
inv_j = np.linalg.inv(j)

print(inv_h)
print(inv_j)
```
